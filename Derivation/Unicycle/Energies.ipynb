# Energy

To apply lagrangian mechanics, we need to calculate the kinetic energy and potential energy of the system.

## Keinetic Energy

The kinetic energy $T$ of the system can be expressed as the sum of the kinetic energies of the wheel and the pendulum.
$$
T_{\text{wheel}} = \frac{1}{2} m_w ||\dot{\vec{X}}_w||^2 
$$

$$
T_{\text{pendulum}} = \frac{1}{2} m_p ||\dot{\vec{X}}_p||^2
$$

Here the trickest part is to calculate $\dot{\vec{X}}_w$ and $\dot{\vec{X}}_p$, who are very long and complicated. We can directly used those in the shifted frame, which should make it a bit easier to calculate.

For the wheel, we have:
$$
\begin{align}
\vec{X}_w &= R \vec{z_2} + R \varphi \vec{x_1},
\end{align}
$$

Thus:
$$
\begin{align}
\dot{\vec{X}}_w &= R \dot{\vec{z_2}} + R \dot{\varphi} \vec{x_1} + R \varphi \dot{\vec{x_1}}
\end{align}
$$



We can first clarify the axis of the frame shifting, which is:
$$
\begin{align}\omega_{f_1} &= \dot{\psi} \vec{z_0}, \\
\omega_{f_2} &= \dot{\theta} \vec{x_1}, \\
\omega_{f_3} &= \dot{\gamma} \vec{y_2},
\end{align}
$$

Then we can write the derivative of $\vec{x_1}$ and $\vec{z_2}$ in the shifted frame as:

$$
\begin{align}
\dot{\vec{x_1}} &= \omega_{f_1} \times \vec{x_1} \\
                &= \dot{\psi} \vec{y_1}  \\
\dot{\vec{z_2}} &=  (\omega_{f_1} + \omega_{f_2}) \times \vec{z_2} \\
                &= \dot{\psi} \vec{z_0} \times \vec{z_2} + \dot{\theta} \vec{x_2} \times \vec{z_2}\\
                &= \dot{\psi} \sin{\theta} \vec{x_1} - \dot{\theta} \vec{y_2} \\
\end{align}
$$

Thus $\dot{\vec{X}}_w$ can be calculated as:
$$
\begin{align}
\dot{\vec{X}}_w &= R \dot{\vec{z_2}} + R \dot{\varphi} \vec{x_1} + R \varphi \dot{\vec{x_1}} \\
                &= R (\dot{\psi} \sin{\theta} \vec{x_1} - \dot{\theta} \vec{y_2}) + R \dot{\varphi} \vec{x_1} + R \varphi \dot{\psi} \vec{y_1} \\
                &= R (\dot{\psi} \sin{\theta} + \dot{\varphi}) \vec{x_1} + R \varphi \dot{\psi} \vec{y_1} - R \dot{\theta} \vec{y_2} \\
                &= R (\dot{\psi} \sin{\theta} + \dot{\varphi}) \vec{x_1} + R \varphi \dot{\psi} \vec{y_1} - R \dot{\theta} \cos{\theta} \vec{y_1} - R \dot{\theta} \sin{\theta} \vec{z_1} \\
                &= R (\dot{\psi} \sin{\theta} + \dot{\varphi}) \vec{x_1} + R (\varphi \dot{\psi} - \dot{\theta} \cos{\theta}) \vec{y_1} - R \dot{\theta} \sin{\theta} \vec{z_1}
\end{align}
$$

And the kinetic energy of the wheel can be calculated as:
$$
\begin{align}
T_{\text{wheel}} &= \frac{1}{2} m_w ||\dot{\vec{X}}_w||^2 \\
                &= \frac{1}{2} m_w R^2 [(\dot{\psi} \sin{\theta} + \dot{\varphi})^2 + (\varphi \dot{\psi} - \dot{\theta} \cos{\theta})^2 + (\dot{\theta} \sin{\theta})^2]           
\end{align}
$$

For the pendulum, we have:
$$
\begin{align}
\vec{X}_p &= \vec{X}_w + h \vec{z_3} = R \vec{z_2} + h \vec{z_3} + R \varphi \vec{x_1},
\end{align}
$$

Similarly we can calculate $\dot{\vec{X}}_p$ as:
$$
\begin{align}
\dot{\vec{X}}_p &= \dot{\vec{X}}_w + h \dot{\vec{z_3}} \\
                &= R (\dot{\psi} \sin{\theta} + \dot{\varphi}) \vec{x_1} + R (\varphi \dot{\psi} - \dot{\theta} \cos{\theta}) \vec{y_1} + R \dot{\theta} \sin{\theta} \vec{z_1} + h \dot{\vec{z_3}}
\end{align}
$$

To calculate $\dot{\vec{z_3}}$, we can also use the shifted frame, and to make it easier we can directly calculate the derivative with $\vec{z_3} = \sin \gamma \vec{x_2} + \cos \gamma \vec{z_2}$ 

$$
\begin{align}
\dot{\vec{z_3}} &= \dot{\gamma} \cos \gamma \vec{x_2} - \dot{\gamma} \sin \gamma \vec{z_2} + \sin \gamma \dot{\vec{x_2}} + \cos \gamma \dot{\vec{z_2}} \\
                &= \dot{\gamma} \cos \gamma \vec{x_2} - \dot{\gamma} \sin \gamma \vec{z_2} + \sin \gamma (\omega_{f_1} + \omega_{f_2}) \times \vec{x_2} + \cos \gamma (\omega_{f_1} + \omega_{f_2}) \times \vec{z_2} \\
\end{align}
$$

For $\dot{\vec{x_2}}$ and $\dot{\vec{z_2}}$, we have:
$$
\begin{align}
\dot{\vec{x_2}} &= \omega_{f_1} \times \vec{x_2} + \omega_{f_2} \times \vec{x_2} \\
                &= \dot{\psi} \vec{z_0} \times \vec{x_2} + \dot{\theta} \vec{x_1} \times \vec{x_2} \\
                &= \dot{\psi} \vec{y_1} \\
                &= \dot{\psi} \cos{\theta} \vec{y_2} - \dot{\psi} \sin{\theta} \vec{z_2} \\
\dot{\vec{z_2}} &= \omega_{f_1} \times \vec{z_2} + \omega_{f_2} \times \vec{z_2} \\
                &= \dot{\psi} \vec{z_0} \times \vec{z_2} + \dot{\theta} \vec{x_1} \times \vec{z_2} \\
                &= \dot{\psi} \sin{\theta} \vec{x_1} - \dot{\theta} \vec{y_2} \\
                &=  \dot{\psi} \sin{\theta} \vec{x_2} - \dot{\theta} \vec{y_2} 
                
\end{align}
$$

Pluging these into the equation of $\dot{\vec{z_3}}$, we have:
$$
\begin{align}
\dot{\vec{z_3}} &= \dot{\gamma} \cos \gamma \vec{x_2} - \dot{\gamma} \sin \gamma \vec{z_2} + \sin \gamma (\dot{\psi} \cos{\theta} \vec{y_2} - \dot{\psi} \sin{\theta} \vec{z_2}) + \cos \gamma (\dot{\psi} \sin{\theta} \vec{x_2} - \dot{\theta} \vec{y_2}) \\
                &= (\dot{\gamma} \cos \gamma + \dot{\psi} \sin{\theta} \cos{\gamma}) \vec{x_2} + (\dot{\psi} \sin{\gamma} \cos{\theta} - \dot{\theta} \cos{\gamma}) \vec{y_2} + (-\dot{\gamma} \sin \gamma - \dot{\psi} \sin{\theta} \sin{\gamma}) \vec{z_2}
\end{align}
$$

Further plugging $\dot{\vec{z_3}}$ into the equation of $\dot{\vec{X}}_p$, we have:
$$
\begin{align}
\dot{\vec{X}}_p &= R (\dot{\psi} \sin{\theta} + \dot{\varphi}) \vec{x_1} + R (\varphi \dot{\psi} - \dot{\theta} \cos{\theta}) \vec{y_1} + R \dot{\theta} \sin{\theta} \vec{z_1} + h \dot{\vec{z_3}} \\
                &= R (\dot{\psi} \sin{\theta} + \dot{\varphi}) \vec{x_1} + R (\varphi \dot{\psi} - \dot{\theta} \cos{\theta}) \vec{y_1} + R \dot{\theta} \sin{\theta} \vec{z_1} + h [(\dot{\gamma} \cos \gamma \\
                & + \dot{\psi} \sin{\theta} \cos{\gamma}) \vec{x_2} + (\dot{\psi} \sin{\gamma} \cos{\theta} - \dot{\theta} \cos{\gamma}) \vec{y_2} \\
                & + (-\dot{\gamma} \sin \gamma - \dot{\psi} \sin{\theta} \sin{\gamma}) \vec{z_2}]
\end{align}
$$

In frame $E_2$, we have:

$$

\dot{\vec{X}}_p =
\begin{bmatrix}
R (\dot{\psi} \sin\theta + \dot{\varphi})
+ h (\dot{\gamma} \cos\gamma + \dot{\psi} \sin\theta \cos\gamma) \\[6pt]

R \varphi \dot{\psi} \cos\theta
- R \dot{\theta} \cos 2\theta
+ h (\dot{\psi} \sin\gamma \cos\theta - \dot{\theta} \cos\gamma) \\[6pt]

- R \varphi \dot{\psi} \sin\theta
+ R \dot{\theta} \sin 2\theta
+ h (-\dot{\gamma} \sin\gamma - \dot{\psi} \sin\theta \sin\gamma)
\end{bmatrix}_{E_2}
$$

Thus the kinetic energy of the pendulum can be calculated as:
$$
\begin{align}
T_{\text{pendulum}} &= \frac{1}{2} m_p ||\dot{\vec{X}}_p||^2 \\
                &= \frac{1}{2} m_p \{[R (\dot{\psi} \sin\theta + \dot{\varphi}) + h (\dot{\gamma} \cos\gamma + \dot{\psi} \sin\theta \cos\gamma)]^2 \\
                & + [R \varphi \dot{\psi} \cos\theta - R \dot{\theta} \cos 2\theta + h (\dot{\psi} \sin\gamma \cos\theta - \dot{\theta} \cos\gamma)]^2 \\
                & + [- R \varphi \dot{\psi} \sin\theta + R \dot{\theta} \sin 2\theta + h (-\dot{\gamma} \sin\gamma - \dot{\psi} \sin\theta \sin\gamma)]^2\}
\end{align}
$$

### Total Kinetic Energy
The total kinetic energy of the system can be expressed as:
$$
\begin{align}
T &= T_{\text{wheel}} + T_{\text{pendulum}} \\
&= \frac{1}{2} m_w R^2 [(\dot{\psi} \sin{\theta} + \dot{\varphi})^2 + (\varphi \dot{\psi} - \dot{\theta} \cos{\theta})^2 + (\dot{\theta} \sin{\theta})^2] \\
& + \frac{1}{2} m_p \{[R (\dot{\psi} \sin\theta + \dot{\varphi}) + h (\dot{\gamma} \cos\gamma + \dot{\psi} \sin\theta \cos\gamma)]^2 \\ 
& + [R \varphi \dot{\psi} \cos\theta - R \dot{\theta} \cos 2\theta + h (\dot{\psi} \sin\gamma \cos\theta - \dot{\theta} \cos\gamma)]^2 \\
& + [- R \varphi \dot{\psi} \sin\theta + R \dot{\theta} \sin 2\theta + h (-\dot{\gamma} \sin\gamma - \dot{\psi} \sin\theta \sin\gamma)]^2\}
\end{align}
$$

## Potential Energy

Potential energy is much easier to calculate, which is only related to the height of the center of mass of the pendulum and the wheel and we don't need to consider derivatives.

Since we already have:
$$
\begin{align}
\vec{X}_w &= R \vec{z_2} + R \varphi \vec{x_1}, \\
\vec{X}_p &= R \vec{z_2} + h \vec{z_3} + R \varphi \vec{x_1},
\end{align}
$$

The potential energy can be calculated as:
$$
\begin{align}
V &= m_w g \vec{X}_w \cdot \vec{z_0} + m_p g \vec{X}_p \cdot \vec{z_0} \\
\end{align} 
$$

For the wheel, we have:
$$
\begin{align}
\vec{X}_w \cdot \vec{z_0} &= (R \vec{z_2} + R \varphi \vec{x_1}) \cdot \vec{z_0} \\
&= R \vec{z_2} \cdot \vec{z_0} + R \varphi \vec{x_1} \cdot \vec{z_0} \\
&= R \cos{\theta}
\end{align}
$$

For the pendulum, we have:
$$
\begin{align}
\vec{X}_p \cdot \vec{z_0} &= (R \vec{z_2} + h \vec{z_3} + R \varphi \vec{x_1}) \cdot \vec{z_0} \\
&= R \vec{z_2} \cdot \vec{z_0} + h \vec{z_3} \cdot \vec{z_0} + R \varphi \vec{x_1} \cdot \vec{z_0} 
\end{align}
$$

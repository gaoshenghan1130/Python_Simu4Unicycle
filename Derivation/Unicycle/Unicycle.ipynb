# Unicycle equation and simulations derivation from Main Equations

The main equations of the system(contraints):

$$
\begin{align}
\omega_1 &:= \dot{\vartheta}, \\
\omega_2 &:= \dot{\varphi} + \dot{\psi} \sin \vartheta, \\
\omega_3 &:= \dot{\psi} \cos \vartheta, \\
\sigma_r &:= \dot{r} - \dot{\vartheta} R, \\
\sigma_\gamma &:= \dot{\gamma}h + \dot{\varphi} R \cos \gamma + \dot{\psi} (h + R \cos \gamma) \sin \vartheta,
\end{align}
$$

$$
\begin{align}
\dot{x}_C &= \omega_1 R \sin \psi \cos \vartheta + \omega_2 R \cos \psi, \\
\dot{y}_C &= -\omega_1 R \cos \psi \cos \vartheta + \omega_2 R \sin \psi, \\
\dot{\psi} &= \frac{\omega_3}{\cos \vartheta}, \\
\dot{\vartheta} &= \omega_1, \\
\dot{\varphi} &= \omega_2 - \omega_3 \tan \vartheta, \\
\dot{r} &= \sigma_r + \omega_1 R, \\
\dot{\gamma} &= \sigma_\gamma  \frac{1}{h} - \omega_2 \frac{R}{h} \cos \gamma - \omega_3 \tan \vartheta,
\end{align}
$$

We also needs to derive $\vec{X}_w = (x_c, y_c, z_c)$ and $\vec{X}_p = (x_p, y_p, z_p)$ for the kinetic energy calculation. The position of the center of mass of the wheel and the pendulum can be expressed as:
$$
\begin{align}
\vec{X}_w &= R \vec{z_2} + R \varphi \vec{x_1}, \\
\vec{X}_p &= \vec{X}_w + h \vec{z_3} = R \vec{z_2} + h \vec{z_3} + R \varphi \vec{x_1},
\end{align}
$$

To simplify $E_n$ axis:

- $E_0 = (x_0, y_0, z_0)$ is the standard basis of the inertial frame, which is fixed to the ground.
- $E_1 = (x_1, y_1, z_1)$ is transformed by rotation $\psi$ from $z_0$, which is fixed to the wheel center.
$$
\begin{align}
x_1 &= \cos \psi x_0 + \sin \psi y_0, \\
y_1 &= -\sin \psi x_0 + \cos \psi y_0, \\
z_1 &= z_0,
\end{align}
$$
- $E_2 = (x_2, y_2, z_2)$ is transformed by rotation $\vartheta$ from $x_1$, which is fixed to the pendulum center.
$$
\begin{align}
x_2 &= x_1, \\
y_2 &= \cos \vartheta y_1 + \sin \vartheta z_1, \\
z_2 &= -\sin \vartheta y_1 + \cos \vartheta z_1,
\end{align}
$$
- $E_3 = (x_3, y_3, z_3)$ is transformed by rotation $\gamma$ from $y_2$, which is fixed to the pendulum center of mass.
$$
\begin{align}
x_3 &= \cos \gamma x_2 - \sin \gamma y_2, \\
y_3 &= y_2, \\
z_3 &= \sin \gamma x_2 + \cos \gamma y_2,
\end{align}
$$

Since we only cares about $z_2$ and $z_3$, we can directly calculate them as:
$$
\begin{align}
z_2 &= -\sin \vartheta y_1 + \cos \vartheta z_1, \\
    &= -\sin \vartheta (-\sin \psi x_0 + \cos \psi y_0) + \cos \vartheta z_0, \\
\end{align}
$$
$$
\begin{align}
z_3 &= \sin \gamma x_2 + \cos \gamma y_2 = \sin \gamma x_1 + \cos \gamma (\cos \vartheta y_1 + \sin \vartheta z_1), \\
    &= \sin \gamma (\cos \psi x_0 + \sin \psi y_0) + \cos \gamma (\cos \vartheta (-\sin \psi x_0 + \cos \psi y_0) + \sin \vartheta z_0), \\
\end{align}
$$  

As a result we have in vector form:

$$
\vec{X}_w = \begin{bmatrix} 
R \sin \vartheta \sin \psi \\ 
-R \sin \vartheta \cos \psi \\ 
R \cos \vartheta 
\end{bmatrix}_{E_0} + R \varphi \begin{bmatrix} 
\cos \psi \\ 
\sin \psi \\
0
\end{bmatrix}_{E_0} = \begin{bmatrix} 
R \sin \vartheta \sin \psi + R \varphi \cos \psi \\ 
-R \sin \vartheta \cos \psi + R \varphi \sin \psi \\ 
R \cos \vartheta

\end{bmatrix}_{E_0}
$$

$$
\begin{align}
\vec{X}_p & = \begin{bmatrix} 
R \sin \vartheta \sin \psi + h (\sin \gamma \cos \psi - \cos \gamma \cos \vartheta \sin \psi) \\ 
-R \sin \vartheta \cos \psi + h (\sin \gamma \sin \psi + \cos \gamma \cos \vartheta \cos \psi) \\ 
R \cos \vartheta + h \cos \gamma \sin \vartheta 
\end{bmatrix}_{E_0} + R \varphi \begin{bmatrix} 
\cos \psi \\ 
\sin \psi \\
0
\end{bmatrix}_{E_0} \\
 &= \begin{bmatrix} 
R \sin \vartheta \sin \psi + h (\sin \gamma \cos \psi - \cos \gamma \cos \vartheta \sin \psi) + R \varphi \cos \psi \\ 
-R \sin \vartheta \cos \psi + h (\sin \gamma \sin \psi + \cos \gamma \cos \vartheta \cos \psi) + R \varphi \sin \psi \\ 
R \cos \vartheta + h \cos \gamma \sin \vartheta
\end{bmatrix}_{E_0}
\end{align}
$$

# Forces and Torques

Here we first need to identify the state variables, basically the params here should include:
- $\psi$ : the heading angle of the segway
- $\vartheta$ : the pitch angle of the segway
- $\varphi$ : the velocity of the wheel rotation, relative to the ground
- $r$ : the position linear motor
- $\gamma$ : the angle between the pendulum and the segway body

Then we can write down the forces and torques action by lagrangian mechanics, and then we can get the equations of motion.

To make the derivation easier, we consider everything as a point mass, 

## Lagrangian Mechanics

### Generalized coordinates and velocities
The generalized coordinates are:
- $q_1 = \psi$ (heading angle)
- $q_2 = \vartheta$ (pitch angle)
- $q_3 = \varphi$ (wheel rotation angle)
- $q_4 = r$ (linear motor position)
- $q_5 = \gamma$ (pendulum angle)

Thus $q = [\psi, \vartheta, \varphi, r, \gamma]^\top$.

## Kinetic Energy
The kinetic energy $T$ of the system can be expressed as the sum of the kinetic energies of the wheel and the pendulum.
$$
T_{\text{wheel}} = \frac{1}{2} m_w ||\dot{\vec{X}}_w||^2 
$$

$$
T_{\text{pendulum}} = \frac{1}{2} m_p ||\dot{\vec{X}}_p||^2
$$

Here the trickest part is to calculate $\dot{\vec{X}}_w$ and $\dot{\vec{X}}_p$, who are very long and complicated. We can directly used those in the shifted frame, which should make it a bit easier to calculate.

For the wheel, we have:
$$
\begin{align}
\vec{X}_w &= R \vec{z_2} + R \varphi \vec{x_1},
\end{align}
$$

Thus:
$$
\begin{align}
\dot{\vec{X}}_w &= R \dot{\vec{z_2}} + R \dot{\varphi} \vec{x_1} + R \varphi \dot{\vec{x_1}}
\end{align}
$$



We can first clarify the axis of the frame shifting, which is:
$$
\begin{align}\omega_{f_1} &= \dot{\psi} \vec{z_0}, \\
\omega_{f_2} &= \dot{\theta} \vec{x_1}, \\
\omega_{f_3} &= \dot{\gamma} \vec{y_2},
\end{align}
$$

Then we can write the derivative of $\vec{x_1}$ and $\vec{z_2}$ in the shifted frame as:

$$
\begin{align}
\dot{\vec{x_1}} &= \omega_{f_1} \times \vec{x_1} \\
                &= \dot{\psi} \vec{y_1}  \\
\dot{\vec{z_2}} &=  (\omega_{f_1} + \omega_{f_2}) \times \vec{z_2} \\
                &= \dot{\psi} \vec{z_0} \times \vec{z_2} + \dot{\theta} \vec{x_2} \times \vec{z_2}\\
                &= \dot{\psi} \sin{\theta} \vec{x_1} - \dot{\theta} \vec{y_2} \\
\end{align}
$$

Thus $\dot{\vec{X}}_w$ can be calculated as:
$$
\begin{align}
\dot{\vec{X}}_w &= R \dot{\vec{z_2}} + R \dot{\varphi} \vec{x_1} + R \varphi \dot{\vec{x_1}} \\
                &= R (\dot{\psi} \sin{\theta} \vec{x_1} - \dot{\theta} \vec{y_2}) + R \dot{\varphi} \vec{x_1} + R \varphi \dot{\psi} \vec{y_1} \\
                &= R (\dot{\psi} \sin{\theta} + \dot{\varphi}) \vec{x_1} + R \varphi \dot{\psi} \vec{y_1} - R \dot{\theta} \vec{y_2} \\
                &= R (\dot{\psi} \sin{\theta} + \dot{\varphi}) \vec{x_1} + R \varphi \dot{\psi} \vec{y_1} - R \dot{\theta} \cos{\theta} \vec{y_1} + R \dot{\theta} \sin{\theta} \vec{z_1}
                &= R (\dot{\psi} \sin{\theta} + \dot{\varphi}) \vec{x_1} + R (\varphi \dot{\psi} - \dot{\theta} \cos{\theta}) \vec{y_1} + R \dot{\theta} \sin{\theta} \vec{z_1}
\end{align}
$$

And the kinetic energy of the wheel can be calculated as:
$$
\begin{align}
T_{\text{wheel}} &= \frac{1}{2} m_w ||\dot{\vec{X}}_w||^2 \\
                &= \frac{1}{2} m_w R^2 [(\dot{\psi} \sin{\theta} + \dot{\varphi})^2 + (\varphi \dot{\psi} - \dot{\theta} \cos{\theta})^2 + (\dot{\theta} \sin{\theta})^2]           
\end{align}
$$
